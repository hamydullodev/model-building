```python
from sklearn.preprocessing import MinMaxScaler
def scaling_qil(df):
    scaler=MinMaxScaler()
    num_col=df.select_dtypes(include=['int32','int64']).columns.drop('condition')
    df[num_col]=scaler.fit_transform(df[num_col])
    return df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def encodla(df):
    encoder=LabelEncoder()
    for col in df.columns:
        if df[col].dtype=='object':
            if df[col].nunique()<=5:
                dummies=pd.get_dummies(df[col],prefix=col, dtype=int)
                df=pd.concat([df.drop(columns=[col]),dummies],axis=1)
            else:
                df[col]=encoder.fit_transform(df[col])
    return df
```

```python
df=encodla(df)
```

```python
def tozala(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=='object':
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(),inplace=True)
    return df
```

```python
from sklearn.preprocessing import MinMaxScaler

def scaling(df):
    num_col = df.select_dtypes(include=['int64', 'float64']).columns.drop(['condition'])
    df[num_col] = MinMaxScaler().fit_transform(df[num_col])
    return df
```

```python
df = scaling(df)
print(df.head())
```

```python
x=df.drop('condition',axis=1)
```

```python
y=df['condition']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test = train_test_split(x, test_size=0.3)
y_train, y_test = train_test_split(y, test_size=0.3)
```

```python
from sklearn.svm import SVC,SVR
```

```python
svc=SVC(kernel='linear',C=1.0)
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```
