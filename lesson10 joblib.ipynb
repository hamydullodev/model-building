```python
import pandas as pd
df=pd.read_csv('heart_cleveland_upload.csv')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
def encodla(df):
    encoder=LabelEncoder()
    for col in df.columns:
        if df[col].dtype=='object':
            if df[col].nunique()<5:
                dummies=pd.get_dummies(df[col],prefix=col,dtype=int)
                df=pd.concat([df.drop(columns=[col]),dummies],axis=1)
            else:
                df[col]=encoder.fit_transform(df[col])
    return df
```

```python
df=encodla(df)
```

```python
from sklearn.preprocessing import MinMaxScaler
def scaling_qil(df):
    scaler=MinMaxScaler()
    num_col=df.select_dtypes(include=['int32','int64']).columns.drop('condition')
    df[num_col]=scaler.fit_transform(df[num_col])
    return df
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

def scaling(df):
    num_col = df.select_dtypes(include=['int64', 'float64']).columns.drop(['condition'])
    df[num_col] = MinMaxScaler().fit_transform(df[num_col])
    return df
```

```python
df = scaling(df)
print(df.head())
```

```python
x=df.drop('condition',axis=1)
```

```python
y=df['condition']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test = train_test_split(x, test_size=0.3)
y_train, y_test = train_test_split(y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
dt_model=DecisionTreeClassifier()
```

```python
dt_model.fit(x_train,y_train)
```

```python
y_pred=dt_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
```

```python
r2=r2_score(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=root_mean_squared_error(y_test,y_pred)
```

```python
print('r2',r2)
print('mae',mae)
print('mse',mse)
print('rmse',rmse)
```
